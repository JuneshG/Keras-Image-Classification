```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import os
```

```python
# create a dataset
data_dir = 'datasets'
```

```python
os.listdir(data_dir)
```

```python
# Remove Dodgy Images
num_skipped = 0
for folder_name in os.listdir(data_dir): #('ThouCat', 'ThouDog'):
    # print(folder_name)
    folder_path = os.path.join(data_dir, folder_name)
    for fname in os.listdir(folder_path):
        # print('fname:',fname)
        fpath = os.path.join(folder_path, fname)
        # print('fpath:',fpath)
        try:
            fobj = open(fpath, "rb")
            is_jfif = tf.compat.as_bytes("JFIF") in fobj.peek(10)
        finally:
            fobj.close()

        if not is_jfif:
            num_skipped += 1
            # Delete corrupted image
            os.remove(fpath)

print("Deleted %d images" % num_skipped)
```

```python
# Create an image generator
image_generator = keras.preprocessing.image.ImageDataGenerator().flow_from_directory(
    data_dir, 
    shuffle=False,
    seed=None,
    follow_links=False,
    batch_size=1,
    target_size=(224, 224), # set the size of the images
    color_mode='rgb',
    class_mode='categorical'
)

# Count the total number of images
total_images = len(image_generator.filenames)

print(f'Total number of images: {total_images}')
```

```python
#create a dataset
dataset = keras.utils.image_dataset_from_directory('newData', batch_size = 50, image_size = (150,150))

for data, labels in dataset:
    print('data shape: ',data.shape)
    print('data type: ',data.dtype)
    print('\n')
    print('labels shape: ',data.shape)
    print('labels dtype: ',data.dtype)
    print('\n')
```

```python
inputs = keras.Input(shape = (150,150,3))
```

```python
from tensorflow.keras import layers
from tensorflow.keras.layers import CenterCrop
from tensorflow.keras.layers import Rescaling
x = CenterCrop(height = 150, width = 150)(inputs)
x = Rescaling(1.0/255.0)(x)

x = layers.Conv2D(filters = 32, kernel_size = (3,3), activation = 'relu')(x)
x = layers.MaxPooling2D(pool_size = (3,3))(x)
x = layers.Conv2D(filters = 32, kernel_size = (3,3), activation = 'relu')(x)
x = layers.MaxPooling2D(pool_size = (3,3))(x)
x = layers.Conv2D(filters= 32, kernel_size = (3,3), activation = 'relu')(x)
x = layers.MaxPooling2D(pool_size = (3,3))(x)

# Apply global average pooling to get flat feature vectors
x = layers.GlobalAveragePooling2D()(x)

#Add a dense classifier on top

num_classes = 2
outputs = layers.Dense(num_classes, activation = 'softmax')(x)

```

```python
model = keras.Model(inputs = inputs, outputs = outputs)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model

# Define the neural network architecture
model = Sequential()
model.add(Dense(64, input_shape=(100,), activation='relu'))
model.add(Dense(10, activation='softmax'))

# Visualize the neural network architecture
plot_model(model, to_file='model.png', show_shapes=True)
```
